# Create Dataframe Using Dictionary

```python
# Impor libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Import csv File

```python
# Import data csv file
df = pd.read_csv('Pokemon.csv')
df
```

```python
# View Pokemon dataset info
df.info()
```

```python
# Checking the null percentage of type 2
round((df.isnull().sum()/len(df)*100),2)
```

```python
# Checking duplicate data
print('Number of duplicated data is:', df.duplicated().sum())
```

```python
# Remove the irrelevant column, i.e. the first column '#'
df = df.drop(['#'], axis=1)
df
```

```python
# Hitunglah jumlah setiap tipe Pokemon(type 1 dan Type 2)
type_1_count = df['Type 1'].value_counts()
type_2_count = df['Type 2'].value_counts().fillna(0)

# Gabungkan data tipe 1 dan tipe 2
print (type_1_count + type_2_count)
# Urutkan tipe berdasarkan jumlah tipe
type_count_sorted = (type_1_count + type_2_count).sort_values(ascending=False)
# Ambil 5 tipe berdasarkan jumlah tipe
top_5_type = type_count_sorted.head(5)
# Cetak informasi jumlah tipe Pokemon
for type, count in top_5_type.items():
  print(f'Type: {type}, jumlah: {count}')
  print()

# Visualisasi data dengan line chart
plt.figure(figsize=(10, 6))
sns.lineplot(x=top_5_type.index, y=top_5_type.values)
plt.title('Jumlah tipe Pokemon')
plt.xlabel('Tipe')
plt.ylabel('Jumlah')
plt.show()
```

```python
 # mengkalkulasi Pokemon tiap generasi
count_per_generation = df['Generation'].value_counts()

# Mengurutkan generasi
sorted_generations = sorted(count_per_generation.index)

# Membuat seri baru dengan generasi terurut
sorted_counts = pd.Series({gen: count_per_generation[gen]for gen in sorted_generations})
print("jumlah Pokemon tiap generasi")
# Melakukan iterasi dari generasi ke generasi secara berurutan
for generation in range(1,7):
  # Filter data untuk generasi saat ini
  generation_data = df[df['Generation'] == generation] # Use 'df' instead of 'data'
  # Menghitung jumlah pokemon di dalam generasi
  count = len(generation_data)
  # Menampilkan jumlah pokemon di dalam generasi
  print(f"Generasi {generation}, terdapat: {count} pokemon")
  print()

# Visualisasikan data dengan bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=sorted_counts.index, y=sorted_counts.values)
plt.title('Jumlah Pokemon tiap generasi')
plt.xlabel('Generasi')
plt.ylabel('Jumlah')  # Menambahkan label sumbu y
plt.show()
```

```python
# Mengelompokan data berdasarkan generasi
generation_data = df.groupby('Generation')

# Pilih statisrik yang ingin dianalisis (HP, Attack, Defense, dll)
statistics = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']

# Group Pokemon by Generation and calculate mean Total stats
mean_stats_by_generation = generation_data[statistics].mean()

# Melt the DataFrame to long format
melted_stats = mean_stats_by_generation.reset_index().melt(id_vars='Generation', var_name='Statistic', value_name='Mean Value') # Reset index to make 'Generation' a column

# Visualisasi data dengan bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Generation', y='Mean Value', hue='Statistic', data=melted_stats)
plt.title('Rata-rata Statistik Pokemon Berdasarkan Generasi')
plt.xlabel('Generasi')
plt.ylabel('Rata-rata')
plt.legend(title='Statistik')
plt.show()
```

```python
# Mencari pokemon terkuat dan terlemah berdasarkan setiap statistik
stats = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
strongest_pokemon = {}
weakest_pokemon = {}

for stat in stats:
    strongest_pokemon[stat] = df.loc[df[stat].idxmax()]
    weakest_pokemon[stat] = df.loc[df[stat].idxmin()]

# Menampilkan informasi Pokemon terkuat dan terlemah
print("Pokemon Terkuat:")
for stat, pokemon in strongest_pokemon.items():
    print(f"{stat}: {pokemon['Name']} ({pokemon[stat]})")

print("\nPokemon Terlemah:")
for stat, pokemon in weakest_pokemon.items():
    print(f"{stat}: {pokemon['Name']} ({pokemon[stat]})")

# Membuat DataFrame untuk visualisasi
df_strongest = pd.DataFrame(strongest_pokemon).transpose()[['Name', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]
df_weakest = pd.DataFrame(weakest_pokemon).transpose()[['Name', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

# Melelehkan DataFrame untuk line chart
df_strongest_melted = df_strongest.melt(id_vars='Name', value_vars=stats, var_name='Stat', value_name='Value')
df_weakest_melted = df_weakest.melt(id_vars='Name', value_vars=stats, var_name='Stat', value_name='Value')

# Visualisasi data dengan bar chart
plt.figure(figsize=(14, 6))
sns.barplot(x='Name', y='Value', hue='Stat', data=df_strongest_melted)
plt.title('Pokemon Terkuat Berdasarkan Statistik')
plt.xlabel('Pokemon')
plt.ylabel('Nilai')
plt.legend(title='Statistik')
plt.show()
```

```python
# Mencari Pokemon terkuat tiap generasi
strongest_pokemon_per_generation = []
for generation in range(1, 7):
    generation_data = df[df['Generation'] == generation]
    if not generation_data.empty:
        strongest_pokemon = generation_data.loc[generation_data['Total'].idxmax()]
        strongest_pokemon_per_generation.append({
            'Generation': generation,
            'Name': strongest_pokemon['Name'],
            'Total': strongest_pokemon['Total']
        })
    else:
        print(f"No Pokemon found for Generation {generation}")

# Menampilkan nama Pokemon terkuat tiap generasi
for pokemon in strongest_pokemon_per_generation:
    print(f"Generasi {pokemon['Generation']}: {pokemon['Name']} (Total: {pokemon['Total']})")

# Visualisasi data dengan line chart
plt.figure(figsize=(10, 6))
sns.lineplot(x=[pokemon['Generation'] for pokemon in strongest_pokemon_per_generation], y=[pokemon['Total'] for pokemon in strongest_pokemon_per_generation])
plt.title('Pokemon Terkuat Tiap Generasi')
plt.xlabel('Generasi')
plt.ylabel('Total')
plt.show()
```

```python
# Mencari pokemon terkuat dan terlemah
strongest_pokemon = df.loc[df['Total'].idxmax()]
weakest_pokemon = df.loc[df['Total'].idxmin()]

# Menampilkan informasi Pokemon terkuat dan terlemah
print("Pokemon Terkuat:")
print(strongest_pokemon[['Name', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Total']])
print("\nPokemon Terlemah:")
print(weakest_pokemon[['Name', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Total']])

# Membuat DataFrame untuk visualisasi
df = pd.DataFrame({
    'Stat': ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'],
    'Terkuat': strongest_pokemon[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].values,
    'Terlemah': weakest_pokemon[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].values
})

# Visualisasi data dengan line chart
plt.figure(figsize=(10, 6))
sns.lineplot(x='Stat', y='Terkuat', data=df, marker='o', label='Terkuat')
sns.lineplot(x='Stat', y='Terlemah', data=df, marker='o', label='Terlemah')
plt.title('Perbandingan Statistik Pokemon Terkuat dan Terlemah (Berdasarkan Total)')
plt.xlabel('Statistik')
plt.ylabel('Nilai')
plt.legend()
plt.show()
```
